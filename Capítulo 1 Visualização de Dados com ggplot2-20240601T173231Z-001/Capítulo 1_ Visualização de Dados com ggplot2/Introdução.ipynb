# **Carros com motores maiores usam mais combustível que carros com motores menores?**

```r
#Data Frame mpg
ggplot2::mpg

#displ = tamannho do motor de um carro, em litros
#hwy = enficiência do combustível de um carro na estrada, em milhas por galão (mpg)
```

```r
#Criando um ggplot
ggplot(data = mpg)+
  geom_point(mapping = aes(x = displ, y = hwy))

#O gráfico mostra uma relação negativa entre o tamanho do motor(displ) e a eficiência do combustível(hwy). Em outras palavras, carros com motores grades usam mais combustível.
```
